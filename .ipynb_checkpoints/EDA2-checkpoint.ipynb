```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
df = pd.read_pickle('df_truncated.pkl')
df.head()
```

```python
df['pricelog'] = np.log(df['price'])
```

```python
df['grg'] = abs(df['long']*df['lat'])
```

```python
df['grglog'] = np.log(df['grg'])
```

```python
f = 'pricelog ~ sqft_living + grade + waterfront + grg + condition'
```
