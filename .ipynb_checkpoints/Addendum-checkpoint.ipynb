```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler     #Feature relative scaling
from sklearn.linear_model import LinearRegression  #OLS driven regression
from sklearn.feature_selection import RFE          #Recursive Feature Elimination
```

```python
#Bring in the cleaned up data, check it
df = pd.read_csv("df_cleaned.csv")
df.head()
```

```python
#Drop columns we won't use for regression analysis
df.drop(['Unnamed: 0', 'id', 'date'], axis=1, inplace=True)
df.columns
```

```python
#Drop columns that we know highly correlate with df.pricelog
df.drop(['sqft_above','sqft_living'], axis=1, inplace=True)
df.columns
```

```python
#Generate pricelog anew
df['pricelog']=np.log(df['price'])
df.head()
```

```python
#Since we're using pricelog we need to also drop price
df.drop('price', axis=1, inplace=True)
```

```python
#Column check
df.columns
```

```python
#Zipcode and Long correlated together (strong negative), Long correlated better with pricelog than zipcode. Dropped zipcode.
df.drop('zipcode', axis=1, inplace=True)
```

```python
#Since we're going to min max scale I need to make sure there isn't anything crazy to throw off the scale
df.describe()
```

```python
#11 bedrooms is a bit much, is this legit?
df.loc[df['bedrooms'] == 11]
```

11 bedrooms is a lot, but it looks like the house sold for $520K w/out a waterfront view, so we know we're looking at a lot of house - plus, it's a 5K squarefoot lot. Doesn't seem too far off.

```python
#Anything else to look into?
df.describe()
```

```python
#start splitting into y and X for scikit learn
y=df.pricelog
X=df.iloc[:,0:15]

#Make sure X copied correctly
X.head()
```

```python
#Define a function that will call scikit learn's MinMaxScaler
scale = MinMaxScaler()

#MinMaxScaler will wipe out our column titles, so we need to preserve them before we transform the dataframe
X_names = X.columns

#Redefining X as the dataframe of the transformed arrays
X = pd.DataFrame(scale.fit_transform(X))

#Test to make sure this was successful
X.head()
```

```python
#Rename column titles by calling the list we stored earlier
X.columns = X_names

#Confirm that this operation took
X.head()
```

```python
#Everything is scaled, but it doesn't make sense to scale the year built or year rennovated
X.yr_built = df.yr_built
X.yr_renovated = df.yr_renovated

#Confirm that it took
X.head()
```

```python
#Just going through here to make sure that the minimum is 0 across the board; the maximum is 1 across the board
X.describe()
```

```python
#Assigning LinearRegression() method to a function
linreg = LinearRegression()

#Nesting RFE traits into a variable called 'selector'
selector = RFE(linreg, n_features_to_select=1, step=1, verbose=True)

#Fitting the selction to X and y
selection = selector.fit(X, y)

#Creates and then displays a dictionary of tuples, where the key is the column name and the value is the rank of importance
#Note 1 is the highest possible rank, 15 is the lowest
ranks = dict(zip(X.columns, selection.ranking_))
print(ranks)
```

```python
#Dataframes are prettier than dictionaries; and they're sortable!
ranks = pd.DataFrame.from_dict(ranks,orient='index', columns=['Rank'])
ranks = ranks.sort_values(by=['Rank'])
ranks
```

```python
#Hard coding the variables to include in SKL's regression
#Included according to rank, so X1 contains the #1 determined coeff
#X2 contains the #2 determined coeff and all coeff specified by X1
#X3 contains the #3 determiend coeff and all coeff specified by X2 ...
X1 = df.loc[:,'grade']
X2 = df.loc[:,['grade','sqft_living15']]
X3 = df.loc[:,['grade','sqft_living15', 'lat']]
X4 = df.loc[:,['grade','sqft_living15', 'lat', 'sqft_lot']]
X5 = df.loc[:,['grade','sqft_living15', 'lat', 'sqft_lot', 'sqft_basement']]
X6 = df.loc[:,['grade','sqft_living15', 'lat', 'sqft_lot', 'sqft_basement', 'waterfront']]
X7 = df.loc[:,['grade','sqft_living15', 'lat', 'sqft_lot', 'sqft_basement', 'waterfront',
                'bathrooms']]
X8 = df.loc[:,['grade','sqft_living15', 'lat', 'sqft_lot', 'sqft_basement', 'waterfront',
                'bathrooms','condition']]
X9 = df.loc[:,['grade','sqft_living15', 'lat', 'sqft_lot', 'sqft_basement', 'waterfront',
                'bathrooms','condition','view']]
X10 = df.loc[:,['grade','sqft_living15', 'lat', 'sqft_lot', 'sqft_basement', 'waterfront',
                'bathrooms','condition','view', 'long']]
X11 = df.loc[:,['grade','sqft_living15', 'lat', 'sqft_lot', 'sqft_basement', 'waterfront',
                'bathrooms','condition','view', 'long', 'bedrooms']]
X12 = df.loc[:,['grade','sqft_living15', 'lat', 'sqft_lot', 'sqft_basement', 'waterfront',
                'bathrooms','condition','view', 'long', 'bedrooms', 'floors']]
X13 = df.loc[:,['grade','sqft_living15', 'lat', 'sqft_lot', 'sqft_basement', 'waterfront',
                'bathrooms','condition','view', 'long', 'bedrooms', 'floors', 'sqft_lot15']]
X14 = df.loc[:,['grade','sqft_living15', 'lat', 'sqft_lot', 'sqft_basement', 'waterfront',
                'bathrooms','condition','view', 'long', 'bedrooms', 'floors', 'sqft_lot15',
                'yr_built']]
X15 = df.loc[:,['grade','sqft_living15', 'lat', 'sqft_lot', 'sqft_basement', 'waterfront',
                'bathrooms','condition','view', 'long', 'bedrooms', 'floors', 'sqft_lot15',
                'yr_built','yr_renovated']]
```

```python
#Since X1 is 1D I have to include a transform step, or linreg won't work
X1 = np.array(X1).reshape(-1,1)

#We're putting them all into a list, because I could put them in a list and define them at the same time
train_sets = [X1, X2, X3, X4,
             X5, X6, X7, X8,
             X9, X10, X11, X12,
             X13, X14, X15]

#Empty variable which will store all the scores of R^2
R_squared=[]

#Model runs linreg on subsets listed in train_sets, then the R^2 of that model is stored in R_squared
for set in train_sets:
    model = linreg.fit(set,y)
    R_squared.append(model.score(set,y))

R_squared
```

```python
#Plotting R^2 versus n variables modeled
plt.figure(figsize=(10,10))
plt.plot(range(1,16), R_squared, marker="o", ls='--', lw=3, ms=10, mfc='red', mec='red')
plt.title('Values of R^2 as n Variables Changes', size=20)
plt.xlabel('n Variables Modeled', size=20)
plt.ylabel('R^2 value returned', size=20);
```
